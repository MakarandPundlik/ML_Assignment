```python
#all import statements here 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([3,4,2,4,5])
x_mean = np.mean(x)
y_mean = np.mean(y)
print(x_mean)
print(y_mean)
```

```python
x_mx = np.array([0,0,0,0,0],dtype=float)
for i in range(0,5):
    x_mx[i] = x[i]-x_mean
    
y_my = np.array([0,0,0,0,0],dtype=float)
for i in range(0,5):
    y_my[i] = y[i]-y_mean
        
print(x_mx)
print(y_my)
```

```python
#square the terms in x-x'
x_mx2 = np.zeros(x_mx.size,dtype=float)
for i in range (0,x_mx.size):
    x_mx2[i] = x_mx[i]*x_mx[i]

print(x_mx2)    
```

```python
#array of x-x' and y-y' product
xy = np.zeros(x_mx.size,dtype=float)
for i in range (0,x_mx.size):
    xy[i] = x_mx[i]*y_my[i]

print(xy) 
```

```python
#formula for m 
sum_m_xy = np.sum(xy)
sum_m_x2 = np.sum(x_mx2)
print(sum_m_xy)
print(sum_m_x2)

m = sum_m_xy/sum_m_x2
print(m)

#get the value of c by putting x and y in eqn 
# y = mx+c 
# c = y - mx
c = y_mean - m*x_mean
print(c)
```

```python
#plot the graph here 
#actual points 
plt.scatter(x, y, color = "m",marker = "o", s = 30)
```

```python
#plotting with predicted value
y_pred = m*x + c
print(y_pred)
plt.plot(x,y_pred,color="g")
plt.ylabel("Y")
plt.xlabel("x")
plt.title("y = 0.4x + 2.4")
```
