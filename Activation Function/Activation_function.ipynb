```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##sigmoid activation function
def sig(x):
    return 1/(1+np.exp(-x)) 
x = np.linspace(-10, 10, 50)   
p = 1/(1+np.exp(-x))
plt.xlabel("x") 
plt.ylabel("Sigmoid(x)")  
plt.title("activation function sigmoid")
plt.plot(x, p) 
plt.show()
```

```python
#tanh activation function
def tan_h(x):
    return np.tanh(x)
ptanh=tan_h(x)
plt.plot(x,ptanh)
plt.title("activation function tanh")
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.show()
```

```python
#reLU activation function
def relu(x):
    return max(0.0,x)
prelu=[relu(u) for u in x]
plt.plot(x,prelu)
plt.title("activation function relu")
plt.xlabel("x")
plt.ylabel("relu(x)")
plt.show()
```

```python
#Leaky ReLU:
def Leaky_relu(a,x):
    return max(a*x,x)
a=0.05
pleaky_relu=[Leaky_relu(a,x) for x in x]
plt.plot(x,pleaky_relu)
plt.title("activation function leaky relu")
plt.xlabel("x")
plt.ylabel("leaky_relu(x)")
plt.grid()
plt.show()
```

```python
#softmax activation function
from scipy.special import softmax
psoftmax=softmax(x)
plt.plot(x,psoftmax)
plt.title("activation function softmax")
plt.xlabel("x")
plt.ylabel("softmax(x)")
plt.grid()
plt.show()
```
