```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt
```

```python
print("enter x coordinates")
x=list(map(int,input().split()))
print("enter y coordinates")
y=list(map(int,input().split()))
```

```python
cord=tuple(zip(x,y)) 
c1=random.choice(cord)
c2=random.choice(cord)
while c2 is c1:
    c2=random.choice(cord)
# c1=(2,3)
# c2=(5,6)
```

```python
d1=[]
d2=[]
for c in cord:
    d1.append(math.sqrt((c1[0]-c[0])**2+(c1[1]-c[1])**2))
    d2.append(math.sqrt((c2[0]-c[0])**2+(c2[1]-c[1])**2))
```

```python
print(d1,d2)
```

```python
cf1=[]
cf2=[]
for i in range(len(d1)):
    if d1[i]>d2[i]:
        cf2.append(cord[i])
    else:
        cf1.append(cord[i])
centroid=[]
centroid.append((np.mean([i[0] for i in cf1]),np.mean([i[1] for i in cf1])))
centroid.append((np.mean([i[0] for i in cf2]),np.mean([i[1] for i in cf2])))
```

```python
plt.scatter([i[0] for i in cf1],[i[1] for i in cf1])
plt.scatter([i[0] for i in cf2],[i[1] for i in cf2])
plt.scatter([i[0] for i in centroid],[i[1] for i in centroid],color='k')
```
